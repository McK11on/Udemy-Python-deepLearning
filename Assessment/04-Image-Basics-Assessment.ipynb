#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2 as cv
import matplotlib.pyplot as plt


import numpy as np
```

```python
path = '../DATA/dog_backpack.jpg'
img = cv.imread(path)
type(img)
```

```python
plt.imshow(img)
```

```python
fix_img = cv.cvtColor(img,cv.COLOR_BGR2RGB,)
plt.imshow(fix_img)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
fix_img = cv.cvtColor(img,cv.COLOR_BGR2RGB,)
```

```python
vertices = np.array([[450,400],[250,700],[650,700]], dtype=np.int32)
pts = vertices.reshape((-1,1,2))
cv.polylines(fix_img, [pts],True, (0,0,255),10)
plt.imshow(fix_img)
```

```python
import cv2 as cv
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros (shape=(512,512,3), dtype= np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv.rectangle(blank_img, (10,10),(300,100), (255,0,0),3)
plt.imshow(blank_img)
```

```python
center_x = blank_img.shape[0]//2
center_y = blank_img.shape[1]//2
```

```python
cv.rectangle(blank_img, (center_x - 50, center_y - 50), (center_x + 50, center_y + 50), (0,0,255), 2)
plt.imshow(blank_img)
```

```python
cv.circle(blank_img, (center_x,center_y), 160, (0,255,0), 4)
plt.imshow(blank_img)
```

```python
cv.circle(blank_img, (450,450), 20, (0,255,0), -1)
plt.imshow(blank_img)
```

```python
cv.line(blank_img, (0,0), (512,512), (102,255,255), 3)
plt.imshow(blank_img)
```

*Note that the layers overlap according to the chosen order of operation

## Text with OpenCv

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(blank_img, 'Hello', (10,480), font, 3, (255,255,255), 3, cv.LINE_AA)
plt.imshow(blank_img)
```

## Custom polygons

```python
blank_img = np.zeros (shape=(512,512,3), dtype= np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,250], [200,200],[400,300],[200,400]], dtype=np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
cv.polylines(blank_img, [pts], True, (255,0,0), 5)
plt.imshow(blank_img)
```
